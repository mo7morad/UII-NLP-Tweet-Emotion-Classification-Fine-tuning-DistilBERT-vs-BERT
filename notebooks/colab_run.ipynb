```python
# Summary of all runs + final test results
import json
import pandas as pd

with open('results/experiments.json') as f:
    store = json.load(f)

rows = []
for run_id, r in store.items():
    rows.append({
        'run': run_id,
        'model': r['model'],
        'lr': r['lr'],
        'batch': r['batch'],
        'epochs': r['epochs'],
        'weighted loss': r['use_weights'],
        'best val macro-F1': round(r['best_val_macro_f1'], 4),
        'test macro-F1': round(r['test_macro_f1'], 4) if 'test_macro_f1' in r else '-',
        'minutes': round(r['runtime_sec'] / 60, 1),
    })
pd.DataFrame(rows)
```
